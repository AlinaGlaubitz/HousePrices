# Data Visualization and Processing -- Categorical Features

1. **Missing Variable Handling:**
Within the categorical data subset, the first step involved addressing missing variables. Instead of imputing missing values, I decided to explicitly designate them as "missing." This approach acknowledges the absence of specific categorical information, preserving the integrity of the data and preventing unintentional biases from imputation.

```python
# For each of the missing values for the qualitative data, we add a new category: MISSING
for c in categorical:
    df_train_houses[c] = df_train_houses[c].astype('category')
    if df_train_houses[c].isnull().any():
        df_train_houses[c] = df_train_houses[c].cat.add_categories(['MISSING'])
        df_train_houses[c] = df_train_houses[c].fillna('MISSING')
```

3. **Translation to Numerical Values:**
Following the treatment of missing variables and the visualization, a crucial step was the translation of categorical values into numerical representations. This process is commonly referred to as encoding. I programmed an encoding that first sorts the categories by the mean sale price and then assigns numerical values based on that. This approach introduces ordinality based on observed relationships with the target variable. By assigning numerical values in alignment with the mean sale prices, the encoding captures the observed hierarchy among categories, potentially enhancing the model's ability to discern patterns and relationships during training.

```python
# This builds an encoder; Problem: does not encode the test data
# How do we solve this problem?
# Possible solutions:
# 1. Work on the test data in the function as well
# 2. Save the dictionaries in a larger dictionary
encodings = dict()
def encode(frame, feature):
    # Data frame is created
    ordering = pd.DataFrame()
    # Add all unique categories for the feature
    ordering['val'] = frame[feature].unique()
    # This becomes the index of the data frame
    ordering.index = ordering.val
    # A new column is added with the mean of the sale price for the category
    ordering['spmean'] = frame[[feature, 'SalePrice']].groupby(feature).mean()['SalePrice']
    # The data frame is ordered by the sale price mean
    ordering = ordering.sort_values('spmean')
    # Each of the categories is assigned a number
    ordering['ordering'] = range(1, ordering.shape[0]+1)
    # These numbers are turned into a dictionary
    ordering = ordering['ordering'].to_dict()
    # For each of the categories, we translate the category in the data set to the value from the dict
    for cat, o in ordering.items():
        frame.loc[frame[feature] == cat, feature+'_E'] = o
    encodings[feature] = ordering

# For the qualitative data, we use the encode function add new columns to the dataframe
cat_encoded = []
for q in categorical:
    encode(df_train_houses, q)
    cat_encoded.append(q+'_E')
print(cat_encoded)
```

ANOVA is employed to assess the statistical significance of the variations in numerical representations across different categories. This shows us the importance of the neighborhood. # Model Building

1. **Train-Test Split:**
The dataset is partitioned into training and validation sets using the train-test split method. This segregation is crucial for model training and subsequent evaluation, ensuring that the model's performance can be assessed on unseen data.
An underlying assumption is made that missing numerical features are absent due to their non-existence in the house.

```python
from sklearn.model_selection import train_test_split

# Split data into train and validation data
X_train, X_val, y_train, y_val = train_test_split(df_train_houses[numerical+cat_encoded].fillna(0.), df_train_houses['SalePrice'], test_size=0.1, random_state=0)
```

2. **Voting Regressor -- Hypertuning for Random Forest:**
Utilizing randomized search, I perform hyperparameter tuning for the Random Forest Regressor. This involves systematically exploring a range of hyperparameter combinations to identify the optimal set that maximizes model performance. Then, we train a model with the best hyperparameters.






```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error as mse
import numpy as np

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)

# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
# Random search of parameters, using 3 fold cross validation,
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 200, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train, y_train)
```

```python
print(rf_random.best_params_)
```

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
from sklearn.metrics import mean_squared_error as mse
# We use the best parameters to create fit a random forest
clf_rf_tuned = RandomForestRegressor(n_estimators=800, min_samples_split=2,
                                      min_samples_leaf=1, max_features='sqrt',
                                      max_depth=40, bootstrap=False,
                                      random_state=42,n_jobs=-1)
clf_rf_tuned.fit(X_train,y_train)
pred_rf_tuned = clf_rf_tuned.predict(X_val)

print('RSME of the (log of) Random Forest with tuned hyperparameters: ',np.sqrt(mse(np.log(pred_rf_tuned),np.log(y_val))))
```

3. **Voting Regressor -- Hypertuning for Gradient Boosting Regressor:**
Similarly, a randomized search is employed to hypertune the Gradient Boosting Regressor. This process seeks to find the most effective hyperparameter configuration for the Gradient Boosting model. Again, a model with the tuned hyperparameters is fit to the data to compare performance. While the Random Forest has a better performance, this regressor is not significantly worse.

```python
from sklearn.ensemble import GradientBoostingRegressor

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(3, 20, num = 18)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]

random_grid = {'n_estimators': n_estimators,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf
              }
print(random_grid)

# Use the random grid to search for best hyperparameters
# First create the base model to tune
gradient = GradientBoostingRegressor()
# Random search of parameters, using 3 fold cross validation,
# search across 100 different combinations, and use all available cores
gradient_random = RandomizedSearchCV(gradient,random_grid, n_iter = 200, cv = 3, verbose=2, n_jobs = -1)
# Fit the random search model
gradient_random.fit(X_train, y_train)
```

```python
print(gradient_random.best_params_)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
# We use the best parameters to create fit a random forest
clf_grad_tuned = GradientBoostingRegressor(n_estimators=1800,min_samples_split=2,min_samples_leaf=2,max_depth=4)
clf_grad_tuned.fit(X_train,y_train)
pred_grad_tuned = clf_grad_tuned.predict(X_val)

print('RSME of the (log of) Gradient Boosting Regressor with tuned hyperparameters: ',np.sqrt(mse(np.log(pred_grad_tuned),np.log(y_val))))
```

4. **Integration into Voting Regressor:**
The tuned Random Forest and Gradient Boosting regressors are integrated into a Voting Regressor. The Voting Regressor combines the predictions of its the two models, incorporating their individual strengths and leveraging an ensemble approach to enhance overall predictive performance. The performance is increased compared to both individual models, even to the better performing Random Forest. This ensemble strategy aims to capture diverse patterns and relationships present in the data, resulting in a more robust and accurate predictive model.

```python
from sklearn.ensemble import VotingRegressor

voting_regressor = VotingRegressor([('rf', clf_rf_tuned), ('grad', clf_grad_tuned)])
voting_regressor.fit(X_train,y_train)
pred_vote = voting_regressor.predict(X_val)

print('Accuracy score of the Voting Regressor with tuned hyperparameters: ',np.sqrt(mse(np.log(pred_vote),np.log(y_val))))
```

**Note on Metric Selection:**
(Metric: RMSE of Logarithm) The chosen evaluation metric is the Root Mean Squared Error (RMSE) of the logarithm of predicted and actual sale prices.
RMSE is a robust metric for regression as it captures the average size of prediction errors, providing a concise measure of overall model accuracy with emphasis on larger errors. Its simplicity and sensitivity make it a widely used and interpretable evaluation criterion. The logarithmic transformation ensures a balanced assessment across the entire price range, preventing undue emphasis on high-priced houses.
